```python
# Install Gradio
!pip install gradio -q

# Import Libraries
import random
import gradio as gr
import matplotlib.pyplot as plt

# -----------------------------------
# PAYOFF LOGIC
# -----------------------------------

def get_scores(move1, move2):

    if move1 == "Cooperate" and move2 == "Cooperate":
        return 3, 3

    elif move1 == "Betray" and move2 == "Cooperate":
        return 5, 0

    elif move1 == "Cooperate" and move2 == "Betray":
        return 0, 5

    else:
        return 1, 1


# -----------------------------------
# AI STRATEGIES
# -----------------------------------

def ai_move(strategy, previous_opponent_move=None):

    if strategy == "Always Cooperate":
        return "Cooperate"

    elif strategy == "Always Betray":
        return "Betray"

    elif strategy == "Random":
        return random.choice(["Cooperate", "Betray"])

    elif strategy == "Tit-for-Tat":

        if previous_opponent_move is None:
            return "Cooperate"

        return previous_opponent_move

    elif strategy == "Mostly Cooperate":
        return random.choices(
            ["Cooperate", "Betray"],
            weights=[80, 20]
        )[0]

    elif strategy == "Mostly Betray":
        return random.choices(
            ["Cooperate", "Betray"],
            weights=[20, 80]
        )[0]


# -----------------------------------
# MAIN SIMULATION FUNCTION
# -----------------------------------

def play_game(player1_strategy, player2_strategy, rounds):

    score1 = 0
    score2 = 0

    history1 = []
    history2 = []

    previous_move1 = None
    previous_move2 = None

    result = ""

    score_progression1 = []
    score_progression2 = []

    cooperation1 = 0
    cooperation2 = 0

    for i in range(rounds):

        move1 = ai_move(player1_strategy, previous_move2)
        move2 = ai_move(player2_strategy, previous_move1)

        previous_move1 = move1
        previous_move2 = move2

        history1.append(move1)
        history2.append(move2)

        if move1 == "Cooperate":
            cooperation1 += 1

        if move2 == "Cooperate":
            cooperation2 += 1

        points1, points2 = get_scores(move1, move2)

        score1 += points1
        score2 += points2

        score_progression1.append(score1)
        score_progression2.append(score2)

        result += f"Round {i+1}\n"
        result += f"Player A: {move1}\n"
        result += f"Player B: {move2}\n"
        result += f"Scores: {score1} - {score2}\n"
        result += "-" * 30 + "\n"

    # Winner Logic
    if score1 > score2:
        winner = "🏆 Player A Wins"

    elif score2 > score1:
        winner = "🏆 Player B Wins"

    else:
        winner = "🤝 It's a Tie"

    result += "\nFINAL RESULTS\n"
    result += "=" * 30 + "\n"
    result += f"Player A Total Score: {score1}\n"
    result += f"Player B Total Score: {score2}\n"
    result += f"{winner}\n\n"

    result += f"Player A Cooperation Count: {cooperation1}\n"
    result += f"Player B Cooperation Count: {cooperation2}\n"

    # -----------------------------------
    # SCORE CHART
    # -----------------------------------

    plt.figure(figsize=(8, 5))

    plt.plot(
        range(1, rounds + 1),
        score_progression1,
        marker='o',
        label="Player A"
    )

    plt.plot(
        range(1, rounds + 1),
        score_progression2,
        marker='o',
        label="Player B"
    )

    plt.xlabel("Rounds")
    plt.ylabel("Scores")
    plt.title("Score Progression")
    plt.legend()
    plt.grid(True)

    chart_path = "score_chart.png"
    plt.savefig(chart_path)
    plt.close()

    return result, chart_path


# -----------------------------------
# GRADIO INTERFACE
# -----------------------------------

app = gr.Interface(
    fn=play_game,

    inputs=[

        gr.Dropdown(
            [
                "Always Cooperate",
                "Always Betray",
                "Random",
                "Tit-for-Tat",
                "Mostly Cooperate",
                "Mostly Betray"
            ],
            label="Player A Strategy"
        ),

        gr.Dropdown(
            [
                "Always Cooperate",
                "Always Betray",
                "Random",
                "Tit-for-Tat",
                "Mostly Cooperate",
                "Mostly Betray"
            ],
            label="Player B Strategy"
        ),

        gr.Slider(
            minimum=1,
            maximum=20,
            value=5,
            step=1,
            label="Number of Rounds"
        )
    ],

    outputs=[
        gr.Textbox(label="Game Results"),
        gr.Image(label="Score Chart")
    ],

    title="🎮 Prisoner's Dilemma Strategy Simulator",

    description="""
Compare different AI strategies in the Prisoner's Dilemma game.
Observe cooperation, betrayal, and score progression.
"""
)

# Launch App
app.launch()
```
